```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
listings_file = "listings.csv"
listings_df = pd.read_csv(listings_file)
listings_df.head()
listings_df.columns
```

```python
listings_clean = listings_df[['id', 'name','property_type', 'listing_url','street','description','amenities','price','security_deposit','cancellation_policy','transit','picture_url','host_response_time','cleaning_fee','extra_people','has_availability']]
listings_clean.head()
```

```python
rds_connection_string = "postgres:1996@localhost:5432/listings_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
listings_clean.to_sql(name='listings', con=engine, if_exists='append', index=False)
```

```python
all_together = pd.read_sql_query('select * from listings', con=engine)
```
