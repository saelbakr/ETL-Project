```python
# dependancies

import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python

# Reading the Csv and creating a df

reviews_file = "Resources/reviews.csv"
reviews_df = pd.read_csv(reviews_file)
reviews_df
```

Transforming the Reviews DataFrame

```python
# selecting only the needed columns for the reviews

reviews_clean = reviews_df[['listing_id','date','comments']]

reviews_clean

# reviews_clean.drop_duplicates(keep=False, inplace=True)
```

Create database connection


```python
# connect to the database

connection= "postgres:postgres@localhost:5432/reviews_db"

# create engine
engine = create_engine(f'postgresql://{connection}')
```

```python
# cheking for the table: reviews

engine.table_names()
```

```python
 Use of  pandas to load csv converted DataFrame into database
```

```python


reviews_clean.to_sql(name='reviews', con=engine, if_exists='append', index=False)
```

```python
# check to see that the table exists in the database

all_reviews = pd.read_sql_query('select * from reviews', con=engine)
```
